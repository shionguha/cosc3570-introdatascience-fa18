```python
import seaborn as sns 
from datascience import *
from datascience.predicates import *
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd 
%matplotlib inline 

plt.style.use('classic')

#Link to csv: https://www.kaggle.com/victorgenin/ices-fish-catch#fish_catches.csv
```

```python
#Table that shows impact of overfishing in Northeast Atlantic Reagon
fish = pd.DataFrame.from_csv("fish_catches.csv")
fish = fish[fish.columns[range(0,12)]]
fish = fish.reset_index()
fish = fish.dropna()
fish
```

<img src='http://www.fao.org/figis/shared/assets/images/maps/fao_fish_area.jpg'>

```python
#Selecting 3 Overfished species 
arraySpecies = ['COD', 'WHB', 'HER']
fish = fish.loc[fish["Species"].isin(arraySpecies)]
fish.head()
```

```python
#Omitting the years with 0's 
#The reason why 0's exist in the data set, is because fish are NOT found in all regions, or NOT Overfished
years = [str(x) for x in range(2006, 2014)]
fish = fish[(fish[years] != 0 ).all(axis=1)]
fish.head()
```

```python
new_data = {"species" : [], "area" : [], "country" : [], "year" : [], "count": []}

for index, row in fish.iterrows():
    for year in range(2006, 2014):
        new_data["species"].append(row["Species"])
        new_data["area"].append(row["Area"])
        new_data["country"].append(row["Country"])
        new_data["year"].append(str(year))
        new_data["count"].append(row[str(year)])

fish = pd.DataFrame(data=new_data)
fish
```

```python
#Frequency diagram 
#Showing how much COD, FlOUNDER, and HADDOCK are present in the data set 
sns.set()

sns.countplot(x = "species", data = fish)
plt.title('Frequency')
```

```python
#Creating scatterplot
sns.set()

g = sns.FacetGrid(fish, col="species")
g.map(plt.scatter, "year", "count")
g.add_legend()
```

```python
#Creating Bar Plot 
ax = sns.FacetGrid(fish, col="species")
ax.map(plt.bar, "year", "count")
g.add_legend()
```
